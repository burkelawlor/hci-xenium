# Proportions

```python
adata = adata_masto.copy()
adata = adata[adata.obs['sample_name'] != 'MAST6']
adata = adata[adata.obs['ct_general'] != 'Unknown/low count']

adata_immune = adata[adata.obs['ct_general'] == 'Immune'].copy()

sample_order = ['NS1', 'NS2', 'NS3', 'NS4', 'MAST1', 'MAST2', 'MAST3']
set_order = ['NL', 'MAST']
```

```python
plot_proportions_bar(
    adata, 'sample_name', 'ct_general', order=sample_order, title='Cell type composition in all cells (general cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=True
)

plot_proportions_bar(
    adata, 'sample_name', 'ct_general',  order=sample_order, title='Cell type composition in all cells (general cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=False,
)


plot_proportions_bar(
    adata, 'sample_set', 'ct_general', order=set_order, title='Cell type compositions pooled',
    display_totals=False, display_totals_title=f'Total cell count', figsize=(6, 2),
)

plot_proportions_bar(
    adata, 'sample_set', 'ct_general', order=set_order, title='Cell type compositions pooled',
    display_totals=False, display_totals_title=f'Total cell count', figsize=(6, 2),
    normalize=False
)
```

```python
plot_proportions_bar(
    adata, 'sample_name', 'ct_detailed', title='Cell type composition in all cells (detailed cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=True, order=sample_order,
)

plot_proportions_bar(
    adata, 'sample_name', 'ct_detailed', title='Cell type composition in all cells (detailed cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=False, order=sample_order,
)

plot_proportions_bar(
    adata, 'sample_set', 'ct_detailed', order=set_order, title='Cell type compositions pooled',
    display_totals=False, display_totals_title=f'Total cell count', figsize=(6, 2),
)

plot_proportions_bar(
    adata, 'sample_set', 'ct_detailed', order=set_order, title='Cell type compositions pooled',
    display_totals=False, display_totals_title=f'Total cell count', figsize=(6, 2),
    normalize=False
)
```
