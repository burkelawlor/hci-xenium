# Load data

```python
adata_lp = ad.read_h5ad("./data/processed/adata/lp+yale_2_annotated.h5ad")

adata = adata_lp.copy()
adata = adata[adata.obs['ct_general'] != 'Unknown/low count']

adata_immune = adata[adata.obs['ct_general'] == 'Immune'].copy()
```

```python
sample_name_order = [
    'NS1','NS2','NS3','NS4',
    'LP1_LP2', 'LP3','LP4','LP5','LP6','LP7',
    'OLP1', 'OLP2', 'OLP3', 'OLP4', 'OLP5', 'OLP6'
    'LPP1', 'LPP2', 'LPP3', 'LPP4',
    'ICI_LP1','ICI_LP2','ICI_LP3','ICI_LP4'
]

sample_set_order = ['NL','LP','OLP','LPP','ICI_LP']
```

# Proportions

```python
plot_proportions_bar(
    adata, 'sample_name', 'ct_general', order=sample_name_order, title='Cell type composition in all cells (general cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=True
)

plot_proportions_bar(
    adata, 'sample_name', 'ct_general',  order=sample_name_order, title='Cell type composition in all cells (general cell types)',
    display_totals=True, display_totals_title=f'Total cell count',
    normalize=False,
)


plot_proportions_bar(
    adata, 'sample_set', 'ct_general', order=sample_set_order, title='Cell type compositions pooled',
    display_totals=True, display_totals_title=f'Total cell count', figsize=(6, 2),
)

plot_proportions_bar(
    adata, 'sample_set', 'ct_general', order=sample_set_order, title='Cell type compositions pooled',
    display_totals=False, display_totals_title=f'Total cell count', figsize=(6, 2),
    normalize=False
)
```
