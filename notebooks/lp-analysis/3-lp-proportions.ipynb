# Load data

```python
adata_lp = ad.read_h5ad("./data/processed/adata/lp_2_annotated.h5ad")
```

```python
sample_name_order = ['NS1','NS2','NS3','NS4','LP1', 'LP2', 'LP3','LPP1', 'LPP2', 'LPP3', 'LPP4','OLP1', 'OLP2', 'OLP3', 'OLP4', 'OLP5', 'OLP6']
sample_set_order = ['NL','LP','LPP','OLP']
```

# Proportions

```python
props = plot_proportions_bar(
    adata_lp, 'sample_name', 'ct_general', order=sample_name_order, title='General cell type composition by sample',
    display_totals=True, return_proportions=True,
)

pct_immune_labels = ['{:,.1f}'.format(v * 100) for v in props['Immune']]
plot_proportions_bar(
    adata_lp[adata_lp.obs['ct_general'] == 'Immune'], 'sample_name', 'ct_immune', order=sample_name_order, title='Immune cell type composition',
    normalize=True, display_totals=pct_immune_labels, display_totals_title=f'% of total cells that are immune',
)
```

```python
props = plot_proportions_bar(
    adata_lp, 'sample_set', 'ct_general', order=sample_set_order, title='General cell type composition by sample set',
    display_totals=True, return_proportions=True,
)

pct_immune_labels = ['{:,.1f}'.format(v * 100) for v in props['Immune']]
plot_proportions_bar(
    adata_lp[adata_lp.obs['ct_general'] == 'Immune'], 'sample_set', 'ct_immune', order=sample_set_order, title='Immune cell type composition by sample set',
    normalize=True, display_totals=pct_immune_labels, display_totals_title=f'% of total cells that are immune',
)
```

```python
props = plot_proportions_line(
    adata_lp[adata_lp.obs['ct_general'] == 'Immune'], 'sample_set', 'ct_immune', 
    order=sample_set_order, title='Immune cell type proportions by sample set',
    sample_name_col='sample_name',
    subset=['CD4+ T cells','CD8+ T cells','Tregs']
)

props = plot_proportions_line(
    adata_lp[adata_lp.obs['ct_general'] == 'Immune'], 'sample_set', 'ct_immune', 
    order=sample_set_order, title='Immune cell type proportions by sample set',
    sample_name_col='sample_name',
    subset=['CLEC10A+ DCs','CLEC9A+ DCs','LAMP3+ DCs','Plasmacytoid DCs']
)
```
